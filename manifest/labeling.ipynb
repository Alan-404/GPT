```python
import pandas as pd
```

```python
df = pd.read_csv("./train.tsv", sep="\t")
```

```python
df
```

```python
diags = []
```

```python
for item in df['input'].to_list():
    data_input = [{
        "text": item, "author": "user"
    }]

    diags.append(data_input)
```

```python
diags
```

```python
pd.DataFrame({
    'input': diags,
    'output': df['output'].to_list()
}).to_csv("diag_train.tsv", sep="\t", index=None)
```
