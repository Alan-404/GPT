```python
import io
import pandas as pd
import numpy as np
import json
import pickle
from sklearn.model_selection import train_test_split

root_folder = "/home/alan/datasets"

dataset = {
    "input": [],
    'output': []
}
dictionary = dict()
actions = []
```

### My Dataset

```python
my_dataset = json.load(open(root_folder + "/chatbot_conversation_vietnamese/train.json", encoding='utf-8'))
```

```python
my_inputs = []
my_outputs = []
```

```python
my_dataset['data']
```

```python
for sample in my_dataset['data']:
    for sample_input in sample['inputs']:
        for sample_output in sample['outputs']:
            my_inputs.append(sample_input)
            my_outputs.append(sample_output)
```

```python
len(my_inputs)
```

```python
len(my_outputs)
```

```python
dataset['input'] += my_inputs
dataset['output'] += my_outputs
dictionary = my_dataset['dictionary']
actions += my_dataset['action']
```

### Open Dataset

```python
open_df = pd.read_csv(f"{root_folder}/Vi-QA/vi-QA.csv")
```

```python
open_df
```

```python
open_df = open_df[(open_df['question'].isna() == False) & (open_df['answers'].isna() == False)]
```

```python
open_df.info()
```

```python
dataset['input'] += open_df['question'].to_list()
dataset['output'] += open_df['answers'].to_list()
```
